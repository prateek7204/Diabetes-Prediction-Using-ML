```python
X = df.drop(columns = 'Outcome', axis = 1)
y = df['Outcome']
```

Doing Scaling of the Data 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
standardized_data
```

```python
X = standardized_data
y = df['Outcome']
```

# Train Test Split Method to split the Dataset into Train Data and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, stratify= y, random_state = 2)
```

```python
X_train
```

# Training SVM and checking Accuracy

```python
from sklearn import svm
model_svm = svm.SVC(kernel='linear')
```

```python
model_svm.fit(X_train,y_train)
```

```python
X_train_pred = model_svm.predict(X_train)
train_data_acc = accuracy_score(X_train_pred, y_train)
print("The Accuracy of training data is : ", train_data_acc)
```

```python
X_test_pred = model_svm.predict(X_test)
test_data_acc = accuracy_score(X_test_pred, y_test)
print("The Accuracy of the Test Data : ", test_data_acc)
```

# Training Logistic Regression and checking Accuracy 

```python
from sklearn.linear_model import LogisticRegression
model_2 = LogisticRegression()
```

```python
model_2.fit(X_train, y_train)
```

```python
X_train_pred = model_2.predict(X_train)
train_data_acc = accuracy_score(X_train_pred, y_train)
print("The accuracy of training data is : ", train_data_acc)
```

```python
X_test_pred = model_2.predict(X_test)
test_data_acc = accuracy_score(X_test_pred,y_test)
print("The Accuracy of the Test Data is : ", test_data_acc)
```

# Training Random Forest and checking Accuracy

```python
from sklearn.ensemble import RandomForestClassifier
model_3 = RandomForestClassifier(n_estimators= 10, criterion="entropy")
```

```python
model_3.fit(X_train, y_train)
```

```python
X_train_pred = model_3.predict(X_train)
train_data_acc = accuracy_score(X_train_pred, y_train)
print("The accuracy of training data is : ", train_data_acc)
```

# GUI (Tkinter)

```python
import tkinter as tk 
from tkinter import messagebox
def predict_diabetes():
    try:
        input_values = [float(entry.get()) for entry in entries]
        np_array = np.asarray(input_values)
        reshaped_array = np_array.reshape(1, -1)
        standard_data = scaler.transform(reshaped_array)
        prediction = model_3.predict(standard_data)
        if prediction[0] == 0:
            result = "The Patient is Non-Diabetic"
        else:
            result = "The Patient is Diabetic"
        messagebox.showinfo("Prediction Result", result)
    except ValueError as e:
        messagebox.showerror("Input Error", f"Invalid input: {e}")
root = tk.Tk()
root.title("Diabetes Prediction")
labels = ["Pregnancies", "Glucose", "Blood Pressure", "Skin Thickness", "Insulin", "BMI", "Diabetes Pedigree Function", "Age"]
entries = []

for label_text in labels:
    label = tk.Label(root, text=label_text)
    label.pack()
    entry = tk.Entry(root)
    entry.pack()
    entries.append(entry)
predict_button = tk.Button(root, text="Predict", command=predict_diabetes)
predict_button.pack()
root.mainloop()
```
